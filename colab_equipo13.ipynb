```python
import sqlite3
import ipywidgets as widgets
from IPython.display import display, clear_output

DB_NAME = "equipo13.db"

def crear_conexion():
    return sqlite3.connect(DB_NAME)

def ejecutar_consulta(query, params=()):
    conn = crear_conexion()
    cursor = conn.cursor()
    cursor.execute(query, params)
    resultados = cursor.fetchall()
    conn.close()
    return resultados

# ---------------- FUNCIONES POR TABLA ----------------
def mostrar_clientes():
    print("\n=== CLIENTES ===")
    for fila in ejecutar_consulta("SELECT id_cliente, nombre, apellido_paterno, direccion FROM clientes"):
        print(f"{fila[0]} - {fila[1]} {fila[2]} | Direcci√≥n: {fila[3]}")

def mostrar_pedidos():
    print("\n=== PEDIDOS ===")
    for fila in ejecutar_consulta("SELECT id_pedido, fecha_pedido, estado_pedido, total_pedido FROM pedido"):
        print(f"#{fila[0]} | Fecha: {fila[1]} | Estado: {fila[2]} | Total: ${fila[3]}")

def mostrar_artistas():
    print("\n=== ARTISTAS ===")
    for fila in ejecutar_consulta("SELECT id_artista, nombre_artista, agencia FROM artista"):
        print(f"{fila[0]} - {fila[1]} | Agencia: {fila[2]}")

def mostrar_productos():
    print("\n=== PRODUCTOS ===")
    for fila in ejecutar_consulta("SELECT id_producto, nombre_producto, precio, stock FROM producto"):
        print(f"{fila[0]} - {fila[1]} | Precio: ${fila[2]} | Stock: {fila[3]}")

def mostrar_pagos():
    print("\n=== PAGOS ===")
    for fila in ejecutar_consulta("SELECT id_pago, fecha_pago, monto_pago, estado_pago FROM pago"):
        print(f"{fila[0]} | Fecha: {fila[1]} | Monto: ${fila[2]} | Estado: {fila[3]}")

def mostrar_metodos_pago():
    print("\n=== M√âTODOS DE PAGO ===")
    for fila in ejecutar_consulta("SELECT id_metodo_pago, tipo_pago FROM metodo_pago"):
        print(f"{fila[0]} - {fila[1]}")

def mostrar_carritos():
    print("\n=== CARRITOS ===")
    for fila in ejecutar_consulta("SELECT id_carrito, id_cliente, fecha_creacion FROM carritos"):
        print(f"{fila[0]} | Cliente ID: {fila[1]} | Creado: {fila[2]}")

def mostrar_carrito_detalle():
    print("\n=== DETALLE DE CARRITOS ===")
    for fila in ejecutar_consulta("SELECT id_carrito_detalle, id_carrito, id_producto, cantidad FROM carrito_detalle"):
        print(f"{fila[0]} | Carrito ID: {fila[1]} | Producto ID: {fila[2]} | Cantidad: {fila[3]}")

def mostrar_envios():
    print("\n=== ENV√çOS ===")
    for fila in ejecutar_consulta("SELECT id_envio, estado_envio, numero_rastreo FROM envio"):
        print(f"{fila[0]} | Estado: {fila[1]} | Rastreo: {fila[2]}")

def mostrar_resenas():
    print("\n=== RESE√ëAS ===")
    for fila in ejecutar_consulta("SELECT id_resena, calificacion, comentario FROM resena"):
        print(f"{fila[0]} | ‚≠ê {fila[1]} - {fila[2]}")

# ---------------- MEN√ö INTERACTIVO ----------------
def menu_interactivo():
    def ejecutar_opcion(opcion):
        clear_output()
        print(f"üìå {opcion}")
        print("-" * 40)
        funciones = {
            "Ver clientes": mostrar_clientes,
            "Ver pedidos": mostrar_pedidos,
            "Ver artistas": mostrar_artistas,
            "Ver productos": mostrar_productos,
            "Ver pagos": mostrar_pagos,
            "Ver m√©todos de pago": mostrar_metodos_pago,
            "Ver carritos": mostrar_carritos,
            "Ver carrito detalle": mostrar_carrito_detalle,
            "Ver env√≠os": mostrar_envios,
            "Ver rese√±as": mostrar_resenas,
            "Salir": lambda: print("üëã ¬°Hasta luego!")
        }

        if opcion in funciones:
            funciones[opcion]()
            if opcion != "Salir":
                display_menu()

    def display_menu():
        opciones = [
            "Ver clientes",
            "Ver pedidos",
            "Ver artistas",
            "Ver productos",
            "Ver pagos",
            "Ver m√©todos de pago",
            "Ver carritos",
            "Ver carrito detalle",
            "Ver env√≠os",
            "Ver rese√±as",
            "Salir"
        ]
        selector = widgets.Dropdown(
            options=opciones,
            description='Selecciona:',
            style={'description_width': 'initial'}
        )
        boton = widgets.Button(description="Ejecutar")

        def on_click(b):
            ejecutar_opcion(selector.value)

        boton.on_click(on_click)
        display(selector, boton)

    display_menu()

# Ejecutar el men√∫
menu_interactivo()
```
